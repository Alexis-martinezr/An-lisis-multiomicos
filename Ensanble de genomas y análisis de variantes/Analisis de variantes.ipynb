```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def pastel(df, x, n1, n2):
    conteo = df[x].value_counts()
    axs[n1, n2].pie(
        conteo,
        labels=conteo.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['blue','yellow','pink','green']
    )
    axs[n1, n2].set_title(f'DistribuciÃ³n de {x}')
```

```python
figure, axs = plt.subplots( 2,2, sharey=False, sharex=False, figsize=(10, 10))
pastel(data, 'IMPACT', 0, 0)
pastel(data, 'Feature_type', 0, 1)
pastel(data, 'Label', 1, 0)
figure.delaxes(axs[1, 1])
plt.tight_layout()
plt.show()
```

```python
count = data.groupby('Consequence')['Consequence'].count().reset_index(name='counts').sort_values(by='counts', ascending=False)

plt.figure(figsize=(15,15))
sns.barplot(data=count, y='Consequence', x='counts', palette='Set2', width=0.9)
plt.tight_layout()
plt.show()
```

```python
count = data.groupby('BIOTYPE')['BIOTYPE'].count().reset_index(name='counts').sort_values(by='counts', ascending=False)

plt.figure(figsize=(12,5))
sns.barplot(data=count, y='BIOTYPE', x='counts', palette='Set2', width=0.9)
plt.tight_layout()
plt.show()
```
